```python
#Arithmetic Operators
a = 10
b = 3
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)
```

```python
#Assignment Operators
a = 10
a += 5
print("After += :", a)
a -= 2
print("After -= :", a)
a *= 3
print("After *= :", a)
a /= 2
print("After /= :", a)
```

```python
#Comparison Operators
a = 15
b = 10
print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

```python
#Logical Operators
a = 20
b = 15
print(a > 10 and b > 10)
print(a > 30 or b > 10)
print(not(a > b))
```

```python
#Identity Operators
a = [1, 2, 3]
b = a
c = [1, 2, 3]
print(a is b)
print(a is c)
print(a is not c)
```

```python
#Membership Operators
fruits = ["apple", "banana", "orange"]
print("apple" in fruits)
print("grapes" in fruits)
print("banana" not in fruits)
```

```python
#Bitwise Operators
a = 5
b = 3
print("AND:", a & b)
print("OR:", a | b)
print("XOR:", a ^ b)
print("Left Shift:", a << 1)
print("Right Shift:", a >> 1)
```

```python
#Operator Precedence
a = 10
b = 5
c = 2
result = a + b * c
print(result)
result = (a + b) * c
print(result)
```

```python
#Check if a number is palindrome (e.g., 121)
num=int(input())
temp=num
rev=0
while num>0:
    dig=num%10
    rev=rev*10+dig
    num=num//10
if rev==temp:
    print("paliandrome num")
else:
    print("non paliandrome num")
```

```python
#Check if a number is Armstrong number (3-digit).
num=int(input())
temp=num
sum=0
while num>0:
    digit=num%10
    sum=sum+digit**3
    num=num//10
if sum==temp:
    print("amstrong number")
else:
    print("not amstrong number")
```

```python
#Check if a number is a perfect square
num = int(input())
i = 1
while i * i <= num:
    if i * i == num:
        print(num, "is a perfect square")
        break
    i += 1
else:
    print(num, "is not a perfect square")
```
