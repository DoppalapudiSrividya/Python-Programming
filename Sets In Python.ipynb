#SETS

```python
#Remove duplicate elements from a list using a set.
numbers = {1, 2, 3, 4, 5, 2, 3, 4}
unique = set(numbers)
unique_list = list(unique)
print(unique_list)
```

```python
#Add an Element
numbers = {1, 2, 3, 4, 5}
numbers.add(6)
print(numbers)
```

```python
#Add Multiple Elements
numbers = {1, 2, 3, 4, 5}
numbers.update([10,20])
print(numbers)
```

```python
#Remove an Element
numbers = {1, 2, 3, 4, 5,6}
numbers.remove(6)
print(numbers)
```

```python
#Remove an Element Safely
numbers = {1, 2, 3, 4, 5,6}
numbers.discard(5)
print(numbers)
```

```python
#Write a program to find the second largest element in a list without using sort() or max().
numbers=(10,20,30,40,50)
largest=numbers[0]
s_largest=numbers[0]
for num in numbers:
    if num>largest:
        s_largest=largest
        largest=num
    elif num > s_largest and num != largest:
        s_largest = num
print(s_largest)
```

```python
#Write a program to count the frequency of each character in a string using a dictionary.
s=input()
count={}
for ch in s:
    if ch in count:
        count[ch]+=1
    else:
        count[ch]=1
print(count) 
```

```python
#Write a program to find whether a given number is an Armstrong Number
num=int(input())
temp=num
sum=0
while num>0:
    digit=num%10
    f=1
    for i in range(1,digit+1):
        f=f*i
    sum=sum+f
    num=num//10
if sum==temp:
    print("Amstrong number")
else:
    print("Not Amstrong number")
```
