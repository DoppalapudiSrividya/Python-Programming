##Dictionaries

```python
#Create a dictionary with your name, age, and city
dict={"name":"srividya","age":"20","city":"tenali"}
print(dict)
```

```python
#Print only the city.
dict={"name":"srividya","age":"20","city":"tenali"}
print(dict["city"])
```

```python
#Add a new key called "course".
dict={"name":"srividya","age":"20","city":"tenali"}
dict["course"]="python"
print(dict)
```

```python
#Update your age.
dict={"name":"srividya","age":"20","city":"tenali"}
dict["age"]=19
print(dict)
```

```python
#Remove the city.
dict={"name":"srividya","age":"20","city":"tenali"}
dict.pop("age")
print(dict)
```

```python
#Print all keys.
dict={"name":"srividya","age":"20","city":"tenali"}
print(dict.keys())
```

```python
#Print all values.
dict={"name":"srividya","age":"20","city":"tenali"}
print(dict.values())
```

```python
#Loop through the dictionary and print each key and value
dict={"name":"srividya","age":"20","city":"tenali"}
for key, value in dict.items():
    print(key, ":", value)
```

```python
#Count the frequency of each character
s=input()
count={}
for ch in s:
    if ch in count:
        count[ch]+=1
    else:
        count[ch]=1
print(count) 
```

```python
#Count the frequency of each word
s=input()
count={}
for word in s.split():
    if word in count:
        count[word]+=1
    else:
        count[word]=1
print(count) 
```

```python
#Find the student with the highest marks
marks = {
    "Rahul": 85,
    "Priya": 92,
    "Kiran": 78,
    "Anu": 95
}
top_student = max(marks, key=marks.get)
print(top_student, marks[top_student])
```

```python
#Merge two dictionaries
d1 = {"a": 1, "b": 2}
d2 = {"c": 3, "d": 4}
d1.update(d2)
print(d1)
```

```python
#Reverse a Dictionary.
d = {
    "a": 1,
    "b": 2,
    "c": 3
}
d1 = {value: key for key, value in d.items()}
print(d1)
```
