## Logical Problems on Strings

```python
#Check whether a string is a palindrome.
s=input()
rev=s[::-1]
if s==rev:
    print("paliandrome")
else:
    print("not paliandrome")
```

```python
#Count the number of uppercase and lowercase letters in a string
s=input()
u_count=0
l_count=0
for ch in s:
    if ch.isupper():
        u_count+=1
    elif ch.islower():
        l_count+=1
print("upper count=",u_count)
print("lower count=",l_count)
```

```python
#Remove all vowels from a string.
s=input()
n=""
for ch in s:
    if ch in"aeiouAEIOU":
        continue
    n+=ch
print(n)
```

```python
#Find the frequency of each character in a string.
s=input()
f = ""
for ch in s:
    if ch not in f:
        print(ch, ":", s.count(ch))
        f += ch
```

```python
#Check whether two strings are anagrams
s=input()
n=input()
if sorted(s)==sorted(n):
    print("anagrame")
else:
    print("not anagrams")
```

```python
#Reverse each word in a sentence without changing the word order.
s=input()
words=s.split()
for word in words:
    print(word[::-1], end=" ")
```

```python
#Find the longest word in a sentence.
s=input()
sub=s.split()
longest=""
for ch in s:
    for word in sub:
        if len(word) > len(longest):
            longest = word
print("Longest word:", longest)
```

```python
#Remove duplicate characters from a string.
s=input()
new=""
for ch in s:
    if ch not in new:
        new+=ch
print(new)
```

```python
#Count the number of digits, alphabets, and special characters in a string
s=input()
digits = 0
alphabets = 0
special = 0
for ch in s:
    if ch.isdigit():
        digits += 1
    elif ch.isalpha():
        alphabets += 1
    else:
        special += 1
print("digits=",digits)
print("alphabets=",alphabets)
print("special=",special)
```
