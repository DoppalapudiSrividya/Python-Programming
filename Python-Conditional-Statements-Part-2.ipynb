```python
#Check Whether a Character is Uppercase or Lowercase
ch=input()
if 65 <= ord(ch) <= 90:
    print("Uppercase")
elif 97 <= ord(ch) <= 122:
    print("Lowercase")
else:
    print("Not an alphabet character")
```

```python
# Check Whether a Character is an Alphabet, Digit, or Special Character
ch=input()
if (65 <= ord(ch) <= 90) or (97 <= ord(ch) <= 122):
    print("Alphabet")
elif 48 <= ord(ch) <= 57:
    print("Digit")
else:
    print("Special Character")
```

```python
# Determine the Type of Triangle from Three Sides
n1=int(input())
n2=int(input())
n3=int(input())
if n1 == n2 == n3:
    print("Equilateral Triangle")
elif n1 == n2 or n2 == n3 or n1 == n3:
    print("Isosceles Triangle")
else:
    print("Scalene Triangle")
```

```python
#  Find the Roots of a Quadratic Equation
a = int(input())
b = int(input())
c = int(input())
d = b*b - 4*a*c   
if d > 0:
    r1 = (-b + d**0.5) / (2*a)
    r2 = (-b - d**0.5) / (2*a)
    print("Two Real Roots:", r1, "and", r2)
elif d == 0:
    r = -b / (2*a)
    print("One Real Root:", r)
else:
    print("Complex Roots")
```

```python
# Shopping Bill Calculator with Discount Slabs
amount=int(input())
if amount >= 5000:
    discount = 0.20   # 20%
elif amount >= 3000:
    discount = 0.15   # 15%
elif amount >= 1000:
    discount = 0.10   # 10%
else:
    discount = 0.05   # 5%
discount_amount = amount * discount
final_amount = amount - discount_amount
print("Total Amount:", amount)
print("Discount Applied:", discount * 100, "%")
print("Discount Value:", discount_amount)
print("Final Payable Amount:", final_amount)
```

```python
units = int(input())
if units <= 100:
    bill = units * 5        
elif units <= 200:
    bill = (100 * 5) + (units - 100) * 7   # ₹7 per unit for next 100
elif units <= 300:
    bill = (100 * 5) + (100 * 7) + (units - 200) * 10  # ₹10 per unit for next 100
else:
    bill = (100 * 5) + (100 * 7) + (100 * 10) + (units - 300) * 15  # ₹15 per unit above 300
print("Total Units:", units)
print("Electricity Bill: ₹", bill)
```

```python
#  Income Tax Calculator Using Tax Slabs
income = float(input("Enter your annual income: "))
if income <= 250000:
    tax = 0
elif income <= 500000:
    tax = (income - 250000) * 0.05
elif income <= 1000000:
    tax = (250000 * 0.05) + (income - 500000) * 0.20
else:
    tax = (250000 * 0.05) + (500000 * 0.20) + (income - 1000000) * 0.30
print("Annual Income:", income)
print("Income Tax Payable:", tax)
```

```python
# Calculate Bonus Based on Employee Experience
experience = int(input("Enter years of experience: "))
salary = float(input("Enter basic salary: "))
if experience >= 10:
    bonus = salary * 0.20   
elif experience >= 5:
    bonus = salary * 0.10   
else:
    bonus = salary * 0.05   
print("Years of Experience:", experience)
print("Basic Salary:", salary)
print("Bonus:", bonus)
print("Total Salary with Bonus:", salary + bonus)
```

```python
# ATM Cash Withdrawal with Minimum Balance Check
balance = int(input("Enter account balance: "))
withdraw = int(input("Enter amount to withdraw: "))
min_balance = 500   
if withdraw <= balance - min_balance:
    balance = balance - withdraw
    print("Withdrawal Successful")
    print("Remaining Balance:", balance)
else:
    print("Transaction Failed: Insufficient Balance")
```

```python
# Simple Calculator Using if-elif-else
a=int(input())
b=int(input())
ch=input("+ - * / % //")
if ch=="+":
    print(a+b)
elif ch=="-":
    print(a-b)
elif ch=="*":
    print(a*b)
elif ch=="/":
    print(a/b)
elif ch=="%":
    print(a%b)
elif ch=="//":
    print(a//b)
else:
    print("invalid choice")
```
